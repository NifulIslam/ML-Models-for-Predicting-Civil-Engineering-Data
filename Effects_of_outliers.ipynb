```python
import numpy as np
from sklearn.feature_selection import mutual_info_regression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_excel('Dataset.xlsx')
data
```

```python
data=data.drop('Soil Type',axis=1)
data
```

## Original data distribution

```python
plt.figure(figsize=(8, 12))  


palette = sns.color_palette("husl", len(data.columns))

for i, column in enumerate(data.columns):
    plt.figure()
    sns.boxplot(y=column, data=data, color=palette[i % len(palette)], orient='h')
    plt.title(f'Horizontal Box Plot of {column}')
    plt.xlabel(column)
    plt.show()
```

## log transform SPT_N

```python
log_transformed_values = np.log(data['SPT_N'])

plt.figure(figsize=(8, 12))

palette = sns.color_palette("husl", len(log_transformed_values))

for i, column in enumerate(['SPT_N']):
    plt.figure()
    sns.boxplot(y=log_transformed_values, color=palette[i % len(palette)], orient='h')
    plt.title(f'Box Plot of Log-Transformed {column}')
    plt.xlabel(f'Log-Transformed {column}')
    plt.show()
```

## Log transform Depth(m)

```python
log_transformed_values = np.log(data['Depth (m)'])

plt.figure(figsize=(8, 12))

palette = sns.color_palette("husl", len(log_transformed_values))

for i, column in enumerate(['Depth (m)']):
    plt.figure()
    sns.boxplot(y=log_transformed_values, color=palette[i % len(palette)], orient='h')
    plt.title(f'Box Plot of Log-Transformed {column}')
    plt.xlabel(f'Log-Transformed {column}')
    plt.show()
```
